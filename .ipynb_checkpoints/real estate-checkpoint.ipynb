```python
# read data
df = pd.read_csv("real-estate.txt", sep="\t", header=0)
df.head()
```

```python
# get lines with missing values -> useless
print(df.isnull().sum())
print(df.count())
```

```python
# in column bathrooms 1 value is missing
# df = df.dropna(axis='columns') # drop whol column ?
df = df.dropna() # or only 1 line
print(df.isnull().sum())
print(df.count())
```

```python
correlation_matrix = df.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True)
```

# conclusion

## convariance matrix result

- price strongly depends on _living area_ and _bedrooms_ (`>.35`)
- price develops divergent to _Rural_, which

```python
#sns.pairplot(df)
sns.pairplot(df, kind="reg")
```

```python
sns.pairplot(df, vars=["Price", "year", "garage"], kind="reg")
```

```python
sns.jointplot(x="year", y="Price", data=df, kind='reg');
```

# conclusion

_Price_ and _year_ have a light correlation, but with a Pearson correlation coefficient of `.13` not a very strong one

```python
sns.jointplot(x="garage", y="Price", data=df, kind='reg')
```

# conclusion

_Price_ and _garage_ have a slightly stronger correlation than _Price_ and _year_.
